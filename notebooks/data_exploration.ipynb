```python
import pandas as pd

cards = pd.read_csv('../data/cards.csv').set_index('cid')
cards.head()
```

```python
decks = pd.read_csv('../data/decks.csv')
decks.head()
```

```python
print('cards:')
cards.isnull().sum()
```

Visualizzaione dei null

```python
cards[cards.source_slug.isnull()]
```

```python
print('decks:')
decks.isnull().sum()
```

```python
d = {}

for ls in decks.cids:
    for x in ls.replace('[', '').replace(']', '').split(','):
        try:
            d[x] += 1
        except:
            d[x] = 1

df_freq = pd.DataFrame({'cids':d.keys(), 'freq':d.values()}).astype('int64')
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(14,6)})
```

```python
sns.histplot(d.values(), kde=True)
```

```python
# top 5 carte più usate
df_freq.nlargest(3, 'freq')
```

```python
import requests
import matplotlib.pyplot as plt
from PIL import Image

def show_cards(cids):
    fig, ax = plt.subplots(figsize=(10,10), nrows=1, ncols=len(cids))
    count_plot = 1
    for cid in cids:
        ax=plt.subplot(1, len(cids), count_plot)
        img = Image.open(requests.get(cards.loc[cid, 'art'], stream=True).raw)
        plt.imshow(img)
        count_plot += 1
        ax.axis('off')
    fig.tight_layout()
```

Carte più utilizzate nei decks

```python
show_cards(df_freq.nlargest(5, 'freq')['cids'])
```

distribuzione potenza media dei decks  

```python
decks.avg_power.hist()
```

```python
d = {}

for ls in decks[decks.avg_power >= 7].cids:
    for x in ls.replace('[', '').replace(']', '').split(','):
        try:
            d[x] += 1
        except:
            d[x] = 1

df_freq_pwf_deck = pd.DataFrame({'cids':d.keys(), 'freq':d.values()}).astype('int64')
```

```python
df_freq_pwf_deck.nlargest(3, 'freq')
```

Carte meta, carte più utilizzate nei decks più potenti

```python
show_cards(df_freq_pwf_deck.nlargest(3, 'freq')['cids'])
```
